## 1. Calculadora Simple:
"""
Crea una función que pueda realizar operaciones básicas como suma, resta, 
multiplicación y división. 
Pedirá al usuario elegir una operación a partir de un listado y luego pedirá los valores a operar.


Utiliza funciones separadas para cada operación.
"""

```python
def sumar(a, b):
    return a + b

def restar(a, b):
    return a - b

def multiplicar(a, b):
    return a * b

def dividir(a, b):
    return a / b

def calculadora():
    print("""
    Elige la operación a realizar:
      1 = Suma
      2 = Resta
      3 = Multiplicación
      4 = División
    """)
    
    a = float(input("Primer número: "))
    b = float(input("Segundo número: "))

    opcion = int(input("Opción: "))

    if opcion < 1 or opcion > 4:
        print("Opción inválida")
        return
    
    if opcion == 4 and b == 0:
        print("Error: división entre cero")
        return
  
    if opcion == 1:
        print("Resultado:", sumar(a, b))
    elif opcion == 2:
        print("Resultado:", restar(a, b))
    elif opcion == 3:
        print("Resultado:", multiplicar(a, b))
    elif opcion == 4:
        print("Resultado:", dividir(a, b))
    else:
        print("Opción inválida")

calculadora()
```

## 2. Número Primo:
"""
Escribe una función que determine si un número dado es primo o no. 
Pedirá al usuario que ingrese un número y muestra un mensaje 
indicando si es primo o no.
"""

```python
def es_primo(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def comprobar_primo():
    num = int(input("Ingresa un número: "))
    if es_primo(num):
        print("Es primo")
    else:
        print("No es primo")

comprobar_primo()
```

## 3. Cálculo del Área:
"""
Implementa funciones para calcular el área de diferentes formas geométricas 
como círculo, cuadrado y triángulo. Pide al usuario que elija la forma y 
luego ingrese los valores necesarios.
"""

```python
import math

def area_circulo(r):
    return math.pi * r**2

def area_cuadrado(l):
    return l * l

def area_triangulo(b, h):
    return (b * h) / 2

def calcular_area():
    print("""
    Opción:
        1 : Círculo
        2 : Cuadrado
        3 : Triángulo
    """)

    opcion = int(input("Elige opción: "))

    if opcion == 1:
        r = float(input("Radio: "))
        print("Área de círculo:", area_circulo(r))
    elif opcion  == 2:
        l = float(input("Lado: "))
        print("Área de cuadrado:", area_cuadrado(l))
    elif opcion == 3:
        b = float(input("Base: "))
        h = float(input("Altura: "))
        print("Área de triángulo:", area_triangulo(b, h))
    else:
        print("Opción inválida")

calcular_area()
```

"""
Crea una función que tome una cadena como entrada y devuelva la cadena invertida. 
Por ejemplo, si la entrada es "python", la salida debería ser "nohtyp".
"""

```python
def invertir_cadena(cadena):
    return cadena[::-1]

def pedir_cadena():
    texto = input("Ingresa una cadena: ")

    print("Original:", texto)
    print("Invertida:", invertir_cadena(texto))

pedir_cadena()
```

```python
def invertir_cadena(cadena):
    invertida = ""
    for c in reversed(cadena):
        invertida += c
    return invertida


def pedir_cadena():
    texto = input("Ingresa una cadena: ")
    print("Original:", texto)
    print("Invertida:", invertir_cadena(texto))


pedir_cadena()
```

```python
def invertir_cadena(cadena):
    partes = cadena.split()      # Separa por espacios: ["2", "5", "65", "39", "2"]
    partes.reverse()             # Invierte la lista
    return " ".join(partes)      # Une de nuevo con espacios


def pedir_cadena():
    texto = input("Ingresa una cadena: ")
    print("Original:", texto)
    print("Invertida:", invertir_cadena(texto))


pedir_cadena()
```

## 6. Fibonacci:
"""
Implementa una función para generar los primeros n números de la 
secuencia de Fibonacci. Pide al usuario que ingrese el valor de n.
"""

```python
def fibonacci(n):
    serie = [0, 1]
    for i in range(2, n):
        serie.append(serie[-1] + serie[-2])
    return serie[:n]

def pedir_fibonacci():
    n = int(input("Cantidad de números Fibonacci: "))
    print("Fibonacci(", n, "):", fibonacci(n))

pedir_fibonacci()
```

## 7. Ordenar Lista:
"""
Escribe una función que ordene una lista de números de manera ascendente 
o descendente según la elección del usuario.
"""

```python
def ordenar_lista(lista, ascendente=True):
    return sorted(lista, reverse=not ascendente)

def pedir_lista():
    datos = input("Ingresa números separados por space: ")
    lista = [float(x) for x in datos.split()]
    
    orden = input("Ascendente (A) o Descendente (D): ").upper()
    asc = True if orden == "A" else False

    print("Lista original:", lista)
    print("Lista ordenada:", ordenar_lista(lista, asc))

pedir_lista()
```

## 8. Factorial:
"""
Crea una función para calcular el factorial de un número. 
Pide al usuario que ingrese un número y muestra el resultado.
"""

```python
def factorial(n):
    if n < 0:
        return "No existe factorial negativo"
    resultado = 1
    for i in range(1, n + 1):
        resultado *= i
    return resultado

def pedir_factorial():
    n = int(input("Ingresa un número: "))
    print("Factorial (", n, "):", factorial(n))

pedir_factorial()
```

## 9. Conversión de Temperatura:
"""
Implementa funciones para convertir entre Celsius y Fahrenheit. 
Pide al usuario que ingrese la temperatura y la unidad, y luego 
realiza la conversión.
"""

```python
def celsius_a_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_a_celsius(f):
    return (f - 32) * 5/9

def conversion_temperatura():
    temp = float(input("Ingresa la temperatura: "))
    unidad = input("Unidad (C/F): ").upper()

    if unidad == "C":
        print("En Fahrenheit(", temp, "):", celsius_a_fahrenheit(temp))
    elif unidad == "F":
        print("En Celsius(", temp, "):", fahrenheit_a_celsius(temp))
    else:
        print("Unidad inválida")

conversion_temperatura()
```

## 10. Juego de Adivinanzas:
"""
Desarrolla un juego simple en el que el programa elige un número aleatorio 
y el jugador tiene que adivinarlo. 
Proporciona pistas sobre si el número es mayor o menor. 
Utiliza funciones para organizar el código.
"""

```python
import random

def juego_adivinanza():
    numero = random.randint(1, 100)
    intento = 0
    print("Adivina el número entre 1 y 100")

    while True:
        intento += 1
        guess = int(input("Tu intento: "))

        if guess < numero:
            print("El número es mayor")
        elif guess > numero:
            print("El número es menor")
        else:
            print(f"¡Correcto! Lo lograste en {intento} intentos")
            print("El número era:", numero)
            break

juego_adivinanza()
```
