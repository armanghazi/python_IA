```python
import re
from PyPDF2 import PdfReader
```

```python
files = [
    "339741.pdf",
    "FDS 2023 - SALFUMAN - AGUA FUERTE.pdf",
    "SDB-9277-ES-ES.pdf"
]

for file in files:
    reader = PdfReader(file)
    
    for page in reader.pages:
        text = page.extract_text()
        
        if text:  # avoid None
            matches = re.findall(r'\bH.{3}\b', text)
            print(f"{file} -> {matches}")
```

```python
all_matches = []

for file in files:
    reader = PdfReader(file)
    for page in reader.pages:
        text = page.extract_text()
        if text:
            all_matches.extend(re.findall(r'\bH.{3}\b', text))

print(all_matches)
```
